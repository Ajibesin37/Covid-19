Select countries for bar plot for all set of countries ans some data preprocessing 

```python
cols = ['date','Afghanistan','Argentina','Australia','United Kingdom','Spain','China','Canada','Russia','US','Venezuela']

subsetdf = data_corona[cols] ## select on above countries 

subsetdf.set_index("date", inplace = True) ## set the column as index column of df 

cum_sum_df =subsetdf.cumsum(axis = 0) ## take cumulative sum of all col

cum_sum_df.tail(10)
```
